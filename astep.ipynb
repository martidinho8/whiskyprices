```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Because why not ;)

pd.options.display.max_rows = 9999 
```

```python
# I got this csv by pasting an html table object into excel (lol-eth) then saving as csv. obviously I want to use
# Beautiful Soup to scrape that element out 

df = pd.read_csv('whisky-a.csv')
```

```python
# Define letter url origin and the date it was published the first time

df['page_letter'] = 'A'
df['price_date'] = '06/12/2008'

```

```python
# Create a numbered index for joining the whisky category which fell off onto the succeeding row

df['C'] = np.arange(len(df))
```

```python
# For the purpose of joining the whisky category name with the previous row

df['cat_index'] = df["C"] - 1
```

```python
# Let's get rows where there is only the whisky category name

cat_df = df[df.More.isnull()]
```

```python
cat_df.head(5)
```

```python
cat_df = cat_df[['cat_index', 'Product Information']]
```

```python
# Just looking at the df that just has category names

cat_df.head(5)
```

```python
# gentle rename action

cat_df = cat_df.rename(columns={'cat_index': 'C', 'Product Information': 'Category'})

```

```python
# First join of the day, let's get the category label to join with the bottle it belonged to 

df_with_cat = pd.merge(df, cat_df, on='C', how='left')

```

```python
# Any product that did not join onto a category label can now leave the building

df_with_cat = df_with_cat[df_with_cat.Category.notnull()]

```

```python
# just dropping some cols in one of the many diff ways in pd

df_with_cat = df_with_cat.drop(df_with_cat.columns[[2, 3, 5, 6]], axis=1)
```

```python
df_with_cat = df_with_cat[['Product Information', 'Category', 'Price(Ex VAT)', 'price_date']]
```

```python
# ok that seems to have worked let's create another index for working out vintages and ages

df_with_cat['join_index'] = np.arange(len(df_with_cat))
```

```python
# I have created a separate filtered df for joining. This will only include bottles with an age statement

df_with_cat_years = df_with_cat[df_with_cat['Product Information'].str.contains("Year old")==True]

df_with_cat_years
df_with_cat['age_string'] = df_with_cat['Product Information'].str.split('Year old').str[0]

df_with_cat
```

```python
df_with_cat_years['age'] = df_with_cat['age_string'].str[-4:]
```

```python
df_with_cat_years['age'] = df_with_cat_years['age'].replace('/', '', regex=True)
```

```python
df_with_cat_years['age'] = df_with_cat_years['age'].replace(' ', '', regex=True)
```

```python
df_with_cat_years['age'] = df_with_cat_years['age'].replace('r', '', regex=True)
```

```python
# After all that cleaning I now have an age column with just numbers to represent years

df_with_cat_years
```

```python
# So now I am going to do something similar with stripping out the bottlesize in cl's
```

```python
df_with_cat['cl_string'] = df_with_cat['Product Information'].str.split('cl / ').str[0]
```

```python
df_with_cat['cl'] = df_with_cat['cl_string'].str[-4:]
```

```python
df_with_cat['cl'] = df_with_cat['cl'].str.extract('(\d+)', expand=False)
```

```python
df_with_cat = df_with_cat[['Product Information', 'Category', 'Price(Ex VAT)', 'cl', 'price_date','join_index']]
```

```python
# ok after a bit of reshuffling and neatening up I am now going to do the same with abv's.

# A PROBLEM HERE IS the ABV GETS ROUNDED DOWN AND LOSES IT'S FLOATINESS, MUST FIX
```

```python
df_with_cat['abv_string'] = df_with_cat['Product Information'].str.split('% / ').str[0]
```

```python
df_with_cat['abv'] = df_with_cat['abv_string'].str[-5:]
```

```python
df_with_cat
```

```python
df_with_cat['abv'] = df_with_cat['abv'].replace('%', '', regex=True)

df_with_cat
```

```python
df_with_cat['abv'] = df_with_cat['abv'].replace('/', '', regex=True)

df_with_cat
```

```python
df_with_cat['abv'] = pd.to_numeric(df_with_cat['abv'],errors='coerce')

df_with_cat.info()
```

```python
df_with_cat['abv'] = df_with_cat['abv'].fillna(0)
```

```python
# not needed anymore?

# df_with_cat['abv'] = df_with_cat['abv'].str.extract('(\d+)', expand=False)
```

```python
# Once the cleaning has been done on the abv it is now time to do some joining, let's neaten up the dedicated age df
```

```python
df_with_cat_years = df_with_cat_years[['age','join_index']]
```

```python
# now we join

df_with_cat = pd.merge(df_with_cat, df_with_cat_years, on='join_index', how='left')

df_with_cat
```

```python
df_with_cat = df_with_cat[['Product Information', 'age', 'Category', 'abv', 'cl', 'Price(Ex VAT)', 'price_date','join_index']]
```

```python
# now lets get some vintages

df_with_cat_vintage = df_with_cat
```

```python
df_with_cat_vintage['vintage_string'] = df_with_cat_vintage['Product Information'].str.split('/').str[0]

df_with_cat_vintage
```

```python
df_with_cat_vintage = df_with_cat_vintage[df_with_cat_vintage['vintage_string'].str.contains("19")==True]
```

```python
df_with_cat_vintage
```

```python
df_with_cat_vintage['vintage_string_two'] = df_with_cat_vintage['vintage_string'].str.extract('(\d+)', expand=False)

df_with_cat_vintage
```

```python
df_with_cat_vintage.vintage_string_two = pd.to_numeric(df_with_cat_vintage.vintage_string_two, errors="coerce")

df_with_cat_vintage.info()
```

```python
df_with_cat_vintage = df_with_cat_vintage[df_with_cat_vintage['vintage_string_two'] > 1700]

df_with_cat_vintage
```

```python
df_with_cat_vintage = df_with_cat_vintage.rename(columns={'vintage_string_two': 'vintage'})
```

```python
df_with_cat_vintage = df_with_cat_vintage[['join_index', 'vintage']]
```

```python
df_with_cat = pd.merge(df_with_cat, df_with_cat_vintage, on='join_index', how='left')
```

```python
df_with_cat = df_with_cat[['Product Information', 'age', 'vintage', 'Category', 'abv', 'cl', 'Price(Ex VAT)', 'price_date','join_index']]
```

```python
df_with_cat
```

```python
# to fix - abv's as full floats
# vintages as ints
#  insert distillery name
#  convert price to inc vat and also store excl vat and inc vat as pence ints
```

```python
df_with_cat['vintage'] = df_with_cat['vintage'].fillna(0)

```

```python
df_with_cat['Price(Ex VAT)'] = df_with_cat['Price(Ex VAT)'].replace('£', '', regex=True)

```

```python
df_with_cat = df_with_cat.rename(columns={'Price(Ex VAT)': 'pricebvat'})

```

```python
df_with_cat['pricebvat'] = df_with_cat['pricebvat'].replace(',', '', regex=True)
```

```python
df_with_cat['pricebvat'] = df_with_cat['pricebvat'].astype(float)

```

```python
df_with_cat = df_with_cat.astype({"vintage": int})

df_with_cat.info()
```

```python
df_with_cat['age'] = df_with_cat['age'].fillna(0)
```

```python
df_with_cat.info()
```

```python
df_with_cat = df_with_cat.astype({"age": int})

df_with_cat
```
